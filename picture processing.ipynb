# 图像处理：灰度图 量化  彩色图 采样  

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import cv2


np.set_printoptions(threshold=np.inf)   #方便查看像素点

img=Image.open('D:/Pictures/timg.jpg')
plt.figure("timg")
gray=img.convert('L')        #L (8-bit pixels, black and white)
plt.imshow(gray,cmap='gray')
plt.show()
print (gray.size)  #图片的尺寸
print (gray.mode)  #图片的模式
print (gray.format)  #图片的格式
```

```python
matrix = np.asarray(gray)        #图像转换为矩阵
print(matrix)
```

```python
a,b=matrix.shape[:2]
img=[]
for x in range(a):
    for y in range(b):
        s=int(matrix[x:(x+1),y:(y+1)]/2)
        img.append(s)
down_img1 = np.array((img), np.uint8).reshape(a,b) # 将采样后的图像量化一半，即取值范围为0-128
plt.imshow(down_img1,cmap='gray')
```

```python
matrix1= np.asarray(down_img1)        #图像转换为矩阵
print(matrix1)
```

```python
a,b=matrix1.shape[:2]
img=[]
for x in range(a):
    for y in range(b):
        s=int(matrix1[x:(x+1),y:(y+1)]/2)
        img.append(s)
down_img2 = np.array((img), np.uint8).reshape(a,b) # 将采样后的图像量化一半，即取值范围为0-128
plt.imshow(down_img2,cmap='gray')
```

```python
matrix2= np.asarray(down_img2)        #图像转换为矩阵
print(matrix2)
```

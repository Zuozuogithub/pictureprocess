# 提取一副彩色图像中红色，用HIS模型处理
RGB就是指Red,Green和Blue,一副图像由这三个channel(通道)构成
Gray就是只有灰度值一个channel
HSV即Hue(色调),Saturation(饱和度)和Value(亮度)三个channel

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

```

# RGB一般可以通过几何法来转化成HIS


```python

def color_space_demo(image):  
    plt.figure(figsize=(14,12))

    plt.subplot(221)
    plt.title('原图HSV')
    image = cv2.imread("D:Pictures/paopao.jpg")
    plt.imshow(image)
    image = cv2.resize(cv2.cvtColor(image,cv2.COLOR_BGR2RGB),(200,200))
    plt.subplot(222)
    plt.title('原图RGB')
    plt.imshow(image)
    gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    plt.subplot(223)
    plt.title('灰度图')
    plt.imshow(gray)
    hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
    plt.subplot(224)
    plt.title('hsv')
    plt.imshow( hsv)




color_space_demo(image)
```

```python
def extrace_object_demo2():
    plt.figure(figsize=(14,12))

    src = cv2.imread("D:Pictures/color.jpg")
    hsv = cv2.cvtColor(src, cv2.COLOR_BGR2HSV)
    lower_hsv = np.array([0, 43, 46])
    upper_hsv = np.array([10, 255, 255])
    mask = cv2.inRange(hsv, lowerb=lower_hsv, upperb=upper_hsv)
    dst = cv2.bitwise_and(src, src, mask=mask)
    plt.subplot(212)
    src = cv2.resize(cv2.cvtColor(src,cv2.COLOR_BGR2RGB),(200,200))
    plt.title('原图RGB')
    plt.imshow(src)
    plt.subplot(221)
    plt.title('提取0-10区间红色')
    plt.imshow(dst)
    
    src = cv2.imread("D:Pictures/color.jpg")
    lower_hsv = np.array([156, 43, 46])
    upper_hsv = np.array([180, 255, 255])
    mask = cv2.inRange(hsv, lowerb=lower_hsv, upperb=upper_hsv)
    dst = cv2.bitwise_and(src, src, mask=mask)
    plt.subplot(222)
    plt.title('提取156-180区间红色')
    plt.imshow(dst)
    plt.show()
    
extrace_object_demo2()
```

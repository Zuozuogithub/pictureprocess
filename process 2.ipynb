# 作业二：打开一副低对比度图像，拉伸其图像，直方图均衡

```python
from matplotlib import pyplot as plt
from PIL import Image
import cv2

img=Image.open('D:/Pictures/cc.jpg')
plt.imshow(img,cmap='gray')

#图像直方图
image = cv2.imread('D:/Pictures/cc.jpg')

hist = cv2.calcHist([image],[0],None,[256],[0,256])

plt.figure()#新建一个图像
plt.title("Grayscale Histogram")#图像的标题
plt.xlabel("Bins")#X轴标签
plt.ylabel("# of Pixels")#Y轴标签
plt.plot(hist)#画图
plt.xlim([0,256])#设置x坐标轴范围
plt.show()#显示图像

```

# 我们拉伸一下图像，采用像素矩阵，进行s=c(r**y),r=2,y=0.6,c=1

```python
import numpy as np
matrix = np.asarray(image)        #图像转换为矩阵
s=int(2**0.6)
matrix=matrix*s
image1= Image.fromarray(matrix) #矩阵转换为图像
plt.imshow(image,cmap='gray')
```

# 进行对数变换 s=clog（1+r），c=1，r=7

```python
import math
s=int(math.log(7))
matrix=matrix*s
image2 = Image.fromarray(matrix) #矩阵转换为图像
plt.imshow(image,cmap='gray')
```

# 比较一下三幅图片


```python
plt.figure(figsize=(16, 8))
plt.subplot(221)
plt.title('原图')
plt.imshow(img)
plt.subplot(222)
plt.title('指数变化')
plt.imshow(image1,cmap='gray')
plt.subplot(212)
plt.title('对数变换')
plt.imshow(image2,cmap='gray')
```

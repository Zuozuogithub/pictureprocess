# 作业三：对一副图像加噪声，进行平滑，锐化作用

# 打开一副图像，分别加椒盐噪声，高斯噪声

```python
from matplotlib import pyplot as plt
from PIL import Image
from PIL import ImageFilter
from numpy import *
from scipy import * 
import numpy as np 
import cv2
```

```python
im=Image.open('D:/Pictures/dd.jpg')
im_gas = im.filter(ImageFilter.GaussianBlur(radius=2))   
plt.imshow(im_gas)
plt.title('高斯噪声')
print(im_gas.mode)
```

# 锐化滤波为了突出显示图像的边界和其他的细节，这些锐化是基于一阶导数和二阶导数的。 一阶导数可以产生粗的图像边缘，并广泛的应用于边缘提取，二阶导数对于精细的细节相应更好，常被用于图像增强。 常用的算子为Sobel和Laplacian

```python
plt.figure(figsize=(16, 8))
im_shapen = im.filter(ImageFilter.SHARPEN)   #锐化滤波
plt.subplot(121)
plt.imshow(im_shapen)
plt.title('锐化')
plt.subplot(122)
plt.imshow(im)
plt.title('原图')
```

# 锐化滤波用来处理高斯噪声效果相当好，保留了很多细节，缺点是细节显得比较突兀

# 平滑滤波在PIL中有两种方式：SMOOTH：平滑滤波• SMOOTH_MORE：平滑滤波（程度更深）

```python
plt.figure(figsize=(16, 10))
im_smooth = im.filter(ImageFilter.SMOOTH)   #平滑滤波
im_smore = im.filter(ImageFilter.SMOOTH_MORE)   #平滑滤波
plt.subplot(222)
plt.imshow(im_shapen)
plt.title('锐化')
plt.subplot(221)
plt.imshow(im)
plt.title('原图')
plt.subplot(223)
plt.imshow(im_smooth)
plt.title('平滑')
plt.subplot(224)
plt.imshow(im_smore)
plt.title('更深的平滑')
```

# 平滑滤波对高斯噪声效果也很好，基本上是还原，但是细节上明显不如锐化

```python
img = np.array(Image.open('D:/Pictures/dd.jpg'))

# 随机生成20000个椒盐
rows, cols, dims = img.shape
for i in range(20000):
    x = np.random.randint(0, rows)
    y = np.random.randint(0, cols)
    img[x, y, :] = 255

plt.figure("椒盐噪声")
plt.imshow(img)
plt.axis('off')
plt.show()
```

# 进行一下锐化，平滑，并且进行比较

```python
plt.figure(figsize=(16, 10))
img = Image.fromarray(img)
img_shapen = img.filter(ImageFilter.SHARPEN)   #锐化滤波
img_smooth = img.filter(ImageFilter.SMOOTH)   #平滑滤波
img_smore = img.filter(ImageFilter.SMOOTH_MORE)   #平滑滤波
plt.subplot(222)
plt.imshow(img_shapen)
plt.title('锐化')
plt.subplot(221)
plt.imshow(im)
plt.title('原图')
plt.subplot(223)
plt.imshow(img_smooth)
plt.title('平滑')
plt.subplot(224)
plt.imshow(img_smore)
plt.title('更深的平滑')
```

# 锐化基本上没有效果，平滑可能比锐化更适合处理椒盐噪声
